```python
def plot_linear_svm(svm_clf,xmin,xmax):
  w = svm_clf.coef_[0]
  b = svm_clf.intercept_[0]
  x0 = np.linspace(xmin,xmax,200)
  x1 = -b/w[1] - w[0]/w[1]*x0
  plt.plot(x0,x1,'-k',linewidth=2, label="decision_boundry")
```

```python
plt.scatter(X[:,0][y==0],X[:,1][y==0], label='class 0')
plt.scatter(X[:,0][y==1],X[:,1][y==1], label='class 1')
plot_linear_svm(svm_clf,0,5)
plt.legend()
plt.show()
```

```python
Xs = np.array([[1, 50], [5, 20], [3, 80], [5, 60]]).astype(np.float64)
ys = np.array([0, 0, 1, 1])
```

```python
svm_clf = SVC(kernel='linear')
svm_clf.fit(Xs,ys)
plt.scatter(Xs[:,0][ys==0],Xs[:,1][ys==0], label='class 0')
plt.scatter(Xs[:,0][ys==1],Xs[:,1][ys==1], label='class 1')
plot_linear_svm(svm_clf,0,6)
plt.title("svm without scaling")
plt.legend()
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
X_transform = ss.fit_transform(Xs)

svm_clf = SVC(kernel='linear')
svm_clf.fit(X_transform,ys)
plt.scatter(X_transform[:,0][ys==0],X_transform[:,1][ys==0], label='class 0')
plt.scatter(X_transform[:,0][ys==1],X_transform[:,1][ys==1], label='class 1')
plot_linear_svm(svm_clf,-2,2)
plt.title("SVM with scaling")
plt.legend()
plt.show()
```

# Pipeline in sklearn

```python
from sklearn.pipeline import Pipeline

pipe = Pipeline(steps=[
                       ("scaling",StandardScaler()),
                       ('model',SVC(C=0.5, kernel='linear',random_state=42))
])
```

```python
pipe.fit(Xs,ys)
```

```python
# During Prediction
new_data = [[2,20]]
pipe.predict(new_data)
```

```python
# During Training
ss = StandardScaler()
X_transform = ss.fit_transform(Xs)
svm_clf = SVC(kernel='linear', C=0.5,random_state=42)
svm_clf.fit(X_transform,ys)
```

```python
# During Prediction
new_data = [[2,20]]
new_transformed = ss.transform(new_data)
svm_clf.predict(new_transformed)
```

## Custom Transformers

```python
from sklearn.base import BaseEstimator,TransformerMixin

class CustomTransformer(BaseEstimator,TransformerMixin):
  #define init method if u have something to get as a parameter

  def fit(self,X):
    self.X = X
    self.std = self.X.std()
    self.mean = self.X.mean()
    return self
  
  def transform(self,X):
    return (self.X - self.mean)/self.std
```
